```python
from bing_image_downloader import downloader
import pandas as pd
```

```python
df_=pd.read_csv('instagram_scrapping.csv')
```

```python
df_.head()
```

```python
df_.isnull().sum()
```

```python
df_.dropna(inplace=True)
```

```python
df_['f']=df_['Followers'].apply(lambda s: s[:len(s)-1])
```

```python
df_['length']=df_['f'].apply(lambda s: len(s))
```

```python
df_=df_[df_['length']>0]
```

```python
df_['f']=df_['Followers'].apply(lambda s: int(s[:len(s)-1]))
```

```python
df_=df_.sort_values(by='f',ascending=False)
```

```python
df_.head(10)
```

```python
names=df_['username'].values
names=[n for n in names]
```

```python
len(names)
```

```python
with open('names_dj.txt','w') as f :
    for k in range(len(names)):
        f.write(str(names[k]))
        f.write('\n')
```

```python
lim=1
#Entrer le nombre d'image pour chaque influenceur(euse)
#on a choisi 1 par exemple
```

```python
for a in names:   
    downloader.download(a, limit=lim,  output_dir='dataset', 
    adult_filter_off=True, force_replace=False, timeout=60)
```
